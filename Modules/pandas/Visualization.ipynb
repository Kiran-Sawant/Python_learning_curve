```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing NASDAQ100 ETF Dataset.

nasdaq = data.DataReader('QQQ', data_source='yahoo').drop(columns='Adj Close')
nasdaq
```

# # df.plot() method

```python
# ploting values in 'Open' & 'Close' columns.

nasdaq.plot(y=['Open', 'Close'])
plt.show()
```

# # Bar-Graphs & Pies.

```python
nasdaq['Close'].mean()
```

```python
def average(n):
    if n < 154.47:
        return 'below average'
    elif n > 150 and n < 160:
        return 'average'
    else:
        return 'above average'
```

```python
performance = nasdaq['Close'].apply(average)
performance
```

```python
performance.value_counts().plot(kind='bar')
plt.show()
```

```python
performance.value_counts().plot(kind='barh')
plt.show()
```

```python
performance.value_counts().plot(kind='pie', legend=True)
plt.show()
```
