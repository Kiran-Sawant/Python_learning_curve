```python
import pandas as pd
```

# # Passing a URL to read_csv() method.

```python
# Getting the URL of UMCSI CSV file.
url = 'http://www.sca.isr.umich.edu/files/tbmics.csv'

# Passing the URL to pd.read_csv() method.
umcsi = pd.read_csv(url, index_col=['YYYY', 'Month'])
umcsi.tail(12)
```

# # converting pandas object to python objects.

```python
# converting a DataFrame column to a list.

umcsi['ICS_ALL'].to_list()
```

```python
# converting a DataFrame column to a Dictionary.
# It will take index as dict key.

umcsi['ICS_ALL'].to_dict()
```

```python
# Converting a single column of a DataFrame to another DataFrame.

umcsi['ICS_ALL'].to_frame()
```

# # Exporting .to_csv()

#### Giving the extension while passing the name is necessary.

```python
# Saving the DataFrame as a CSV file locally.

umcsi.to_csv('UMCSI_june-20.csv')
```

# # Importing Excel files

#### While importing excel files giving file extension(.xls, .xlsx, .xlsm, etc) is necessary.
#### Excel files can have multiple work-sheets, use the "sheet_name" parameter to get data from a specific work-sheet, it takes sheet-name or sheet-index as an argument.

```python
# Reading an Excel file with a single sheet.

pd.read_excel('Datasets/Data - Single Worksheet.xlsx')
```

```python
# Reading a single sheet in Excel file with multiple sheets.
# Reading the second sheet of an Excel file using sheet-index.

pd.read_excel('Datasets/Data - Multiple Worksheets.xlsx', sheet_name=1)
```

```python
# Reading the second sheet of an Excel file using sheet-name.

pd.read_excel('Datasets/Data - Multiple Worksheets.xlsx', sheet_name='Data 2')
```

#### # Reading multiple sheets simultaniously from an Excel file will return a dictionary, where the key of the dictionary is the value you pass in the "sheet_name" parameter or the actual sheet name, and the value is a DataFrame.

```python
# Reading specific sheets of an Excel file.

pd.read_excel('Datasets/Data - Multiple Worksheets.xlsx', sheet_name=['Data 1', 'Data 2'])
```

```python
# Extracting the DataFrame from the dictionary.

dataDict = pd.read_excel('Datasets/Data - Multiple Worksheets.xlsx', sheet_name=['Data 1', 'Data 2'])
dataDict['Data 2']
```

```python
# Reading data from all the sheets in an Excel file.

pd.read_excel('Datasets/Data - Multiple Worksheets.xlsx', sheet_name=None)
```

# # Exporting to excel files.

### pd.ExcelWriter() Object

#### Writing to Excel files requires a pandas ExcelWriter object That takes the path\name of the excel file to save in. If the file does not exists it will create a new one and if a file with same name already exists it will overwrite it.
#### The sheets in the Excel file will be created in the same order as you assign them.
#### The ExcelWriter() Object is needed to be saved using the .save() method after all the Data is assigned.

```python
females = dataDict['Data 2'][dataDict['Data 2']['Gender'] == 'F']
females
```

```python
males = dataDict['Data 1'][dataDict['Data 1']['Gender'] == 'M']
males
```

```python
# Creating a pandas Excel Writer object.

excelfile = pd.ExcelWriter('Friends.xlsx')
```

```python
# Writing a DataFrame to an Excel file

females.to_excel(excelfile, sheet_name='Girls', index=False)     # index=False will exclude row-header in Excel file.
```

```python
# Writing only certain columns of a DataFrame to an Excel file

males.to_excel(excelfile, sheet_name='Boys', index=False, columns=['First Name', 'City'])
```

```python
# Saving all the changes.

excelfile.save()
```
