```python
import pandas as pd
import xlwings as xl
```

#### Using pd.read_excel() on an excel sheet with multiple tables is error prone as seen below.
#### Use the xlwings module to specifically target the cell range to be converted to pandas series or dataFrame.

```python
# Reading an excel sheet with multiple tables.

pd.read_excel('Datasets/experiment.xlsx', sheet_name='Brent')
```

```python
# Settingup xlwings Book & Book.sheets object.

book = xl.Book('Datasets/experiment.xlsx')      # Setting Book
brent_sheet = book.sheets('Brent')              # Setting Sheet
```

```python
# Saving a table in Excel as a Dataframe.

brent_frame = brent_sheet.range('A3:E1292').options(pd.DataFrame).value
brent_frame
```

```python
# "index=False" will not set an index.

brent_frame = brent_sheet.range('A3:E1292').options(pd.DataFrame, index=False).value
brent_frame
```

### Here, a Table has a Date column hat does not mention a day that leeds to an error while converting it to timeseries.

```python
# Saving the table as a DataFrame without a timeseries index.

brent_M1 = brent_sheet.range('H3:L63').options(pd.DataFrame, index=False).value
```

```python
# converting the dates to strings & replacing year with day

brent_M1['Date'] = brent_M1['Date'].astype('str').str.replace('2020-', '01-')
```

```python
# Converting the new date strings to pandas Timestamp.

brent_M1['Date'] = pd.to_datetime(brent_M1['Date'], dayfirst=True)
```

```python
# Setting the timeseries column as index

brent_M1.set_index(keys='Date', inplace=True)
```

```python
# Adding a new sheet to Excel file & pasing the modifyed dataframe to it.

brentm1 = book.sheets.add('Brent_m1', after='Brent')
brentm1.range('A3').value = brent_M1
```

```python
book.save()      # Saves The changes to the disk.
```
