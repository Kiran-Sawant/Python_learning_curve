```python
import pandas as pd
import numpy as np
```

```python
# Creating a random Dataset.
data = np.random.randint(0, 100, [1000, 50])

# Converting the dataset to a DataFrame.
df = pd.DataFrame(data)
df
```

# # .max_rows & .min_rows

#### The number of rows that pandas displays when the DataFrame is called can be set & displayed by using the pd.options.display.max_rows or ...min_rows attribute.
#### If a DataFrame has 100 rows and the .max_rows is set to 200, all the rows will be displayed.
#### If you want to view more rows use 'min_rows' which is set to 10 rows by default, it displays 5 rows from the head & 5 rows from the tail and truncates the rest.

```python
# Getting the maximum number of rows that can be displayed.

pd.options.display.max_rows
```

```python
# Setting the maximum number of rows that can be displayed.

pd.options.display.max_rows = 60
```

```python
df
```

```python
# Getting the minimum number of rows that can be displayed.

pd.options.display.min_rows
```

```python
# Setting the minimum number of rows that can be displayed.
# It will display 10 rows from the head & 10 from tail. 

pd.options.display.min_rows = 20
```

```python
df
```

```python
# Getting the maximum number of columns that can be displayed.

pd.options.display.max_columns
```

```python
# Setting the maximum number of columns that can be displayed.

pd.options.display.max_columns = 30
```

```python
df
```

# # pd.get_option() & pd.set_option()

#### The pd.get_option() & pd.set_option() can also be used to display & set the values.
#### The methods take a regular-expression describing the option you want to access and the value you want to assign to it. Use the docString to know all the options one can set & get.

```python
# Getting the maximum number of rows that can be displayed.

pd.get_option('max_rows')
```

```python
# Getting the maximum number of columns that can be displayed.

pd.get_option('max_columns')
```

```python
# Setting the maximum number of rows that can be displayed.

pd.set_option('max_rows', 16)
df
```

# # pd.reset_option()

#### It can be used to reset the options to their default values.

```python
# resetting the max_rows option to default.

pd.reset_option('max_rows')
```

```python
pd.get_option('max_rows')
```

# pd.describe_option()
#### It gives a summary of the option passed as argument.

```python
pd.describe_option('max_rows')
```

# # .precision() Option

#### The .precision option can be used to set the number of decimal places in the floating point numbers in the series or DataFrames.

```python
# Generating a random DataFrame.

df = pd.DataFrame(np.random.randn(5, 5))
df
```

```python
# Getting the precision level of floats.

pd.get_option('precision')
```

```python
# Setting the precision level of floats to 3 decimal places.

pd.set_option('precision', 3)
df
```

```python
# Resetting the precision level to default.
# resetting will return the original dataframe.

pd.reset_option('precision')
df
```
