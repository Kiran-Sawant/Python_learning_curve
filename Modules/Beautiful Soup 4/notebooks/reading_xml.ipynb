```python
import bs4 as bs
import urllib.request as request
```

```python
# Getting the HTML byte string.
sauce = request.urlopen(r"https://pythonprogramming.net/sitemap.xml").read()
# Creating a Soup object from the HTML.
xml_soup = bs.BeautifulSoup(sauce, 'xml')
```

```python
xml_soup.find_all('loc')
```

```python
locs = xml_soup.find_all('loc')

for i in locs:
    print(i.text)
```
