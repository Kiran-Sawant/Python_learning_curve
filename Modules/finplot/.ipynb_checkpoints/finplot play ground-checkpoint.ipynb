```python
import pandas_datareader as pdr
import finplot as fplt
```

```python
rates = pdr.DataReader('GS', data_source='yahoo', start='2020-01-01')
rates
```

```python
fplt.candlestick_ochl(rates[['Open', 'Close', 'High', 'Low']])
fplt.show()
```

```python
rates[['Open', 'Close', 'High', 'Low']]
```

```python
ax = fplt.create_plot(title='Goldman Sachs')
axo = ax.overlay()
fplt.volume_ocv(rates[['Open', 'Close', 'Volume']], ax=axo)
fplt.plot(rates.Volume.ewm(span=24).mean(), ax=axo, color=1)
fplt.candlestick_ochl(rates[['Open', 'Close', 'High', 'Low']])
fplt.show()
```
